```python
def proportion_of_education():

  # Load data
  df = pd.read_csv('NISPUF17.csv')

  # Collect column
  education = df['EDUC1']

  # Filtering education categories
  LSTHS = len(education[education == 1])
  HS = len(education[education == 2])
  MTHSBNC = len(education[education == 3])
  COLLEGE = len(education[education == 4])

  # Finding percentages
  total = len(education)
  a = LSTHS/total
  b = HS/total
  c = MTHSBNC/total
  d = COLLEGE/total

  # Dictionary
  dic = {"less than high school": a,
    "high school": b,
    "more than high school but not college": c,
    "college": d}

  return dic
```

```python
import pprint

dic = proportion_of_education()
pprint.pprint(dic)
```

```python
# Importing libraries
import pandas as pd
import numpy as np

def answer_one():

    # Read file and skip useless rows
    energy = pd.ExcelFile('Energy Indicators.xls')
    energy = pd.read_excel(energy, 0, skiprows=17, skipfooter=38)

    # Catch columns of interest and rename them
    energy = energy[['Unnamed: 1', 'Petajoules', 'Gigajoules', '%']]
    energy = energy.rename(columns={'Unnamed: 1': 'Country',
                                'Petajoules': 'Energy Supply',
                                'Gigajoules': 'Energy Supply per Capita',
                                '%': '% Renewable'})

    # Treating missing values and converting energy in petajoules to gigajoules
    energy['Energy Supply'].replace('...', np.nan, inplace=True)
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000

    # Renaming some values and cleaning the country's name's format
    energy['Country'].replace({'Republic of Korea':'South Korea',
                            'United States of America':'United States',
                            'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
                            'China, Hong Kong Special Administrative Region':'Hong Kong'}, inplace=True)
    energy['Country'].replace(r' \(.*\)', '',regex=True, inplace=True)
    energy['Country'].replace(r' \d', '',regex=True, inplace=True)

    # Reading file and skiping useless rows
    gdp = pd.read_csv('world_bank.csv', skiprows=4)

    # Replacing values
    gdp['Country Name'].replace({"Korea, Rep.": "South Korea",  "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"}, inplace=True)

    # Slicing columns of interest
    gdp = gdp[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    gdp.rename(columns={'Country Name': 'Country'}, inplace=True)

    # Reading file and keeping only the top 15
    ScimEn = pd.read_excel('scimagojr-3.xlsx')
    ScimEn = ScimEn[:15]

    # Joining all three DataFrame
    sci_energy = pd.merge(ScimEn, energy, how='inner', left_on='Country', right_on='Country')
    merged = pd.merge(sci_energy, gdp, how='inner', left_on='Country', right_on='Country')
    merged = merged.set_index('Country')
    return merged

    raise NotImplementedError()
```

```python
def answer_two():

    # First, we need to know the length of all DataFrame objects combined,
    # then, we subtract from the final merged DataFrame

    # Reading all DataFrames and then concatenating them
    # This is just the same first steps of answer_one function

    # Read file and skip useless rows
    energy = pd.ExcelFile('assets/Energy Indicators.xls')
    energy = pd.read_excel(energy, 0, skiprows=17, skipfooter=38)

    # Catch columns of interest and rename them
    energy = energy[['Unnamed: 1', 'Petajoules', 'Gigajoules', '%']]
    energy = energy.rename(columns={'Unnamed: 1': 'Country',
                                'Petajoules': 'Energy Supply',
                                'Gigajoules': 'Energy Supply per Capita',
                                '%': '% Renewable'})

    # Treating missing values and converting energy in petajoules to gigajoules
    energy['Energy Supply'].replace('...', np.nan, inplace=True)
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000

    # Renaming some values and cleaning the country's name's format
    energy['Country'].replace({'Republic of Korea':'South Korea',
                            'United States of America':'United States',
                            'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
                            'China, Hong Kong Special Administrative Region':'Hong Kong'}, inplace=True)
    energy['Country'].replace(r' \(.*\)', '',regex=True, inplace=True)
    energy['Country'].replace(r' \d', '',regex=True, inplace=True)

    # Reading file and skiping useless rows
    gdp = pd.read_csv('assets/world_bank.csv', skiprows=4)

    # Replacing values
    gdp['Country Name'].replace({"Korea, Rep.": "South Korea",  "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"}, inplace=True)

    # Slicing columns of interest
    gdp = gdp[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    gdp.rename(columns={'Country Name': 'Country'}, inplace=True)

    # Reading file and keeping all rows
    ScimEn = pd.read_excel('assets/scimagojr-3.xlsx')

    # Putting everything together
    merged = pd.merge(ScimEn, energy, how='inner',left_on='Country', right_on='Country')
    merged = pd.merge(merged, gdp, how='inner', left_on='Country', right_on='Country')
    merged = merged.set_index('Country')

    concatenated = pd.concat([ScimEn, gdp, energy])['Country'].unique()

    return len(concatenated) - len(merged)
    raise NotImplementedError()
```

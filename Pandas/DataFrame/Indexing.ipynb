# Indexing

The index is essentially a row level label, and in pandas the rows correspond to axis 0. Indices can be either autogenerated or set explicitly. Another option for setting an index is to use the set_index() function, that takes a list of columns and promotes those columns to index.

```python
import pandas as pd

# Create DataFrame
admissions_df = pd.read_csv('dataset/Admission_Predict.csv', index_col=0)

# Clean up columns names
cols = admissions_df.columns
cols = [x.strip() for x in cols]
admissions_df.columns = cols

admissions_df.head()
```

```python
# Changing the DataFrame index from Serial No. to Chance of Admit

# Saving the indexed data to its own (new) column
admissions_df['Serial Number'] = admissions_df.index

# Setting the index to another column
admissions_df = admissions_df.set_index('Chance of Admit')

admissions_df.head()
```

If we want to reset the index, we can do so by using the reset function. It will make the current index a new column and create a default numbered index.

```python
admissions_df = admissions_df.reset_index()
admissions_df.head()
```

## Census dataset

In this dataset there are two summarized levels, one that contains summary data for whole county, and one that contains summary data for each state.

```python
census = pd.read_csv('dataset/census.csv')

census.head()
```

```python
# We see that the possible values for the sum level are using the unique function
# on the DataFrame, which is similar to the SQL distinct operator

# Running unique on the sum level
census['SUMLEV'].unique()
```

```python
# There are only two values. Let's exclude all of the rows that are summaries
# at the state level and just keep the country data.
census = census[census['SUMLEV'] == 50]
census.head()
```

```python
# Reducing the data to just the total population estimates and number of births
target_columns = ['STNAME', 'CTYNAME', 'BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012', 'BIRTHS2013', 'BIRTHS2014', 'BIRTHS2015',
                'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']

census = census[target_columns]
census.head()
```

```python
# Setting the a multi-column index
census = census.set_index(['STNAME', 'CTYNAME'])
census.head()
```

```python
# Getting data for a specific county in a specific state
census.loc['Michigan', 'Alcona County']
```

```python
# Comparing two different counties from different states
# We need to pass the indexes in a tuple
census.loc[ [('Michigan', 'Alcona County'),
             ('Michigan', 'Marquette County')] ]
```

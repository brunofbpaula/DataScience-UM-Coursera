# Data Cleaning

Some Pandas API functions.

```python
import pandas as pd
```

```python
presidents = pd.read_csv('presidents (1).csv')
presidents.head()
```

```python
# Cleaning up with the apply() function
# It takes some arbitrary function we wrote and apply it to
# either a Series (single column) or DataFrame across all rows or columns.

def splitname(row):

  # Extracting the firstname and creating a new entry in the Series
  row['First'] = row['President'].split(' ')[0]

  # Now for the lastname
  row['Last'] = row['President'].split(' ')[-1]

  # Then we return the row and pandas apply() will
  # take of merging them back into a DataFrame
  return row

presidents = presidents.apply(splitname, axis='columns')
presidents.head()
```

```python
# Removing the columns in place
del(presidents['First'])
del(presidents['Last'])
```

```python
# Using regex - Capturing a group
pattern="(^[\w]*)(?:.* )([\w]*$)"

# Extract function built in the str attribute of the Series object
presidents['President'].str.extract(pattern).head()
```

```python
# Refining the pattern to give name to columns
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names = presidents['President'].str.extract(pattern).head()
```

```python
names
```

```python
# Copying the columns into the main DataFrame
presidents['First'] = names['First']
presidents['Last'] = names['Last']

presidents.head()
```

```python
# Now cleaning up the Born column
# Getting rid of anything that's not in the pattern of MM/DD/YYYY
pattern = "([\w]{3} [\w]{1,2}, [\w]{4})"
presidents['Born'] = presidents['Born'].str.extract(pattern)
presidents['Born'].head()
```

```python
# Pandas date/time feature
presidents['Born'] = pd.to_datetime(presidents['Born'])
presidents['Born'].head()
```

# Numpy

#Array Creation
They are displayed as a list or list of lists and can be created through list as well. When creating an arrray, we pass in a list as an argument in numpy array.

```python
a = np.array([1,2,3])
print(f'One-dimension A-array: {a}')
print(f'Dimensions of A-array: {a.ndim}')  # this attribute returns the number of dimensions of a list
print(f'Type of items in A-array: {a.dtype}') # it returns the datatype of the items in the array
```

```python
b = np.array([[4,5,6], [7,8,9]])
print(f'Two-dimension B-array: \n{b}')
print(f'Length of each dimension of B-array: {b.shape}') # it shows the length of each dimension in a array
```

```python
c = np.array([1,2.3,3.4]) # an array can be made up of more than integers.
print(c)
print(c.dtype.name)
```

We can note that numpy will automatically convert intengers up to floats, since there is no loss of precision. Numpy will try and give the best data type format possible to keep your data type homogeneous, which means all the same, in the array.

# Zeros, Ones and Random
Sometimes we know the shape of the array we want to create but not what we want to be in it. Numpy offers several functions to create arrays with initial placeholders, such as zeros or ones.

```python
d = np.zeros((2,3)) # It takes a tuple with the dimensions of the array as argument
print(f'Array of zeros:\n{d}')
e = np.ones((2,3))
print(f'Array of ones:\n{e}')
f = np.random.rand(2,3) # It creates a array with placeholders between 0 and 1 in the given shape
print(f'Array of random numbers:\n{f}')
```

# Sequences
It's also possible to create a sequence of numbers within a range, with the arange() function. The first argument is the starting bound and the second argument is the ending bound, and the third argument is the difference between each consecutive number, just like the range() function.

```python
f = np.arange(2,10+1,2)
print(f)
```

Now, to create a sequence of floats, we use the linspace() function. In this function, the third argument isn't the difference between two numbers, but the total numbers of itens you want to generate. The first and second arguments are inclusive.

```python
g = np.linspace(1,2,5)
print(g)
```

# Matrix Operation
Numpy also supports matrix manipulation.

```python
# Here is an example of matrix product.
a = np.array([[0,1], [1,2]])
b = np.array([[2,4], [2,2]])

print('ELEMENTWISE PRODUCT') # If we want to do elementwise product, we use the '*' sign.
print(a*b)

print('MATRIX PRODUCT') # If we want to do matrix product, we use the '@' sign or the dot function.
print(a@b)
```

```python
# Remembering that the product of two matrices is only plausible when the inner dimensions of the two matrices are the same.
print(f'Shape of the matrix: A({a.shape}, B({b.shape}))')
```

# Images

Arrays can be seen as giant ordered lists of numbers, and the shape of an array as just an abstraction that we have for a particular purpose. This is exactly how basic images are stored in computer enviroment.

```python
from PIL import Image  # Python imaging library
from IPython.display import display  # A function to display images in the Jupyter notebook

# Now, let's open an image
img = Image.open("img/re;.jpeg")
display(img)
```

```python
# We can conver this PIL image to a numpy array
array = np.array(img)
print(array.shape)
print(array, array.dtype)
```

The uint8 datatype means that they are unsigned intengers (so no negative numbers) and the 8 means 8 bites per byte. It means each value could be up to 256 (2^8) in size (actually 255, because we start at zero).

For black and white images, black is stored as 0 and white as 255. There's a way to invert this image using an numpy array:

```python
mask = np.full(array.shape, 255)  # It will create an new array in the shape of the previous one but fulled with 255's.
mask
```

```python
# We subtract that from the modified array
modified_array = array - mask

# We convert all of the negative values to positive values
modified_array = modified_array * -1

# And the last step is to set the value of the datatype correctly
modified_array = modified_array.astype(np.uint8)
modified_array
```

```python
# Lastly, we can display this new array by using the fromarray() function in Python imaging library
# to convert the numpy array into an object jupyter can render
display(Image.fromarray(modified_array))
```

```python
# Now let's try to reshape the original array
reshaped = array.reshape(178, 712, 3)
display(Image.fromarray(reshaped))
```

In Matrix terms, the number of elements always has to equal the product of the number of rows and columns. Don't forget that this is only possible because the new shape of the array is equivalent to the original.

The original size is 356x356x3 that's equal to 380208 elements, and 178x712x3 is also 380208.

# Indexing, Slicing and Iterating

These three are extremely important for data manipulation and analysis because these techiniques allow us to select data based on conditions and copy or update data.

## Indexing

```python
# A one-dimensional array is basically a list. So, to get a specific element, we just need to use the offset index.
z = np.array([1, 2, 3, 4, 5])
z[2]
```

```python
# For multidimensional (two or more) array, we need to use integer array indexing.
y = np.array([[1, 2], [3, 4], [5, 6]])

# If we want to select one certain element, we can do so by entering the index, which is comprised of two
# integers the first being the ROW and second the column. Remembering that we start at zero.
y[2,1]
```

```python
# It's also possible to enter indices directly into an array function
np.array([y[0,0], y[1,1], y[2,1]])
```

```python
# There's another way of doing that by using another form of array indexing,
# which essentially "zips" the first list and the second list up
print(y[[0,1,2], [0,1,1]])
```

## Bolean Indexing
It allows us to select arbitrary elements based on conditions.

```python
y > 5
```

```python
# Another avenue is to place this array of booleans like a mask over the original array to return a one-dimensional
# array relating to the true values
y[y>5]
```

## Slicing
This is a way to create a sub-array based on the original array. For one-dimensional array, slicing works in similar way to a list. To slice, we use the : sign. For instance, if we put :3 in the indexing brackets, we get the elements from index 0 to 3 (excluding 3).

```python
w = np.array([0,1,2,3,4,5])
print(w[:3])
w[2:4]
```

```python
v = np.array([[10,11,12], [13,14,15], [15,16,17]])
v
```

```python
# Putting one argument in the array returns all the elements from the first selected row to the last selected row (exclusive)
v[:2]
```

```python
# If we add another argument, for example v[:2, 1:3], we get the first two rows but then the
# second and third columns values only
v[:2, 1:3]
```

It is important to realize that a slice of an array is a view into the same data. This is called passing by reference. Modifying the sub array will modify  the original array as well.


```python
sub_array = v[:2, 1:3]
print(f'Sub array index[0,0] value before change: {sub_array[0,0]}')
sub_array[0,0] = 99
print(f'Sub array index[0,0] value after change: {sub_array[0,0]}')
print(f'Original array index[0,1] value after change: {v[0,1]}')
```
